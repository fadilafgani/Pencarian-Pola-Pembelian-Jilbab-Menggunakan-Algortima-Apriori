```python
import numpy as np
import pandas as pd
import matplotlib as plt
from apyori import apriori 
```

```python
transaksi_jilbab = pd.read_csv('transaksi_jilbab.csv', header=None)
```

```python
transaksi_jilbab.head()
```

```python
transaksi_jilbab.shape
```

```python
records = []
for i in range(0, 11):
    records.append([str(transaksi_jilbab.values[i,j]) for j in range(1, 7)])
```

```python
association_rules = apriori(records, min_support = 0.0045, min_confidence = 0.7, min_lift = 3, min_length = 2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
